## Python With Statement 

Why do we use python with statement ? 
With statement in python is used in exception hangling to make the code cleaner and much more readable. 

Observe the following code example on how to use of with statement makes code cleaner.

# file handling 

1) without using the with statement 

```
file = open("file_path" , 'w')
file.write("helloworld")
file.close()
```

2) without using the with statement 


```
file = open("file_path" , 'w')
try:
  file.write("helloworld")
finally:
  file.close()
```
3) Using the with statement 


```
with open("file_path" , 'w') as file:
      file.write("helloworld")
```


Notice that unlike the first two implementations there is no need to call file.close() when using with statement. The with statement itself ensures proper acquistion and realse of resources. 

The second approach in the above example takes care of all the exceptions but using the with statement makes the code compact and much more readable. 

The with statement is popularly used with file streams.

# Training and Validation datasets

while Building real world machine learning models , it is quite common to split the dataset in to 3parts. 

# Training set :- 
used to train the model that is compute the loss and adjust the weights of the model using gradient descent

# Validation set :- 
used to evaluate the model while training , adjust hyper parameters (learning rate etc) and pick the best version of the model

# Test set :- 
Used to compare different models or different types of modeling approaches and report the final accuracy of the model 

Since there is no pre-defined validation set , we can set aside a small portion of the training set to be used as the validation set. Let's define a function that randomly picks a given fraction of the element indices for creating the validation set. we will also pass a random seed in to the function so we recreate the same training /validation split in the future runs. 

````python
import numpy as np
def split_indices(n , val_pict , random_seed):
  # calculate the validaiton portion
  n_val = int(n*val_pict)
  # set the random seed for reproducability 
  np.random.seed(random_seed)
  # let's add permutuation for the dataset from 0 to n-1
  idxs = np.random.permutation(n)
  return idxs[n_val:] , idxs[:n_val]

validation_percent = 0.2
random_seed = 42
total_number = len(dataset)
train_indices , val_indices = split_indices(n = total_number , val_pict = validation_percent , random_seed = random_seed)
print("The length of the training indices :" , len(train_indices))
print("The length of the validation indices :" , len(val_indices))
print("sample validation indices :" , val_indices[:10])
````

# `Access To GPU `

````python
import torch 
# Get the default device 
def get_default_device():
  """ Pick GPU if available , else CPU """
  if torch.cuda.is_available():
    return torch.device("cuda")
  else:
    return torch.device("cpu")

# moving to device 
def to_device(data , device):
  """ Move Tensors to chosen device """
  if isinstance(data , (list , tuple)):
    return [to_device(x , device) for x in data]
  return data.to(device , non_blocking = True)

# Wrapping the data loader 
class DeviceDataLoader():
  """ Wrap a data loader to move data to a device"""
  def __init__(self , dl , device):
    self.dl = dl
    self.device = device

  def __iter__(self):
    """ Yield a batch of data after moving it to device """ 
    for images , labels in self.dl:
      yield to_device(images , self.device)
      yield to_device(labels , self.device)

  def __len__(self):
    """ Number Of Batches """
    return len(self.dl)

my_device =  get_default_device()
train_dl = DeviceDataLoader(dl = train_loader , device = my_device)
valid_dl = DeviceDataLoader(dl = validation_loader , device = my_device)
````

Before we define the model , let's look at how single convolutional layer followeed by a max-pooling layer operates on the data.

````python
import torch 
def conv_block(in_channels, out_channels, pool=False):
    layers = [torch.nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1), 
              torch.nn.BatchNorm2d(out_channels), 
              torch.nn.ReLU(inplace=True)]
    if pool: layers.append(torch.nn.MaxPool2d(2))
    return torch.nn.Sequential(*layers)

# Let's create this big model 
class Complex_Model(torch.nn.Module):
  def __init__(self , input_channels , number_of_classes):
    # call the parent constructor 
    super(Complex_Model , self).__init__()
    # Input_size = 100 * 3 * 32 * 32 
    self.conv1 = conv_block(input_channels, 64) # 100 * 64 * 32 * 32 
    self.conv2 = conv_block(64, 128, pool=True) # 100 * 128 * 16 *16
    
    self.conv3 = conv_block(128, 256, pool=True) # 100 * 256 * 8 * 8
    self.conv4 = conv_block(256, 512, pool=True) # 100 * 512 * 4 * 4

    self.classifier = torch.nn.Sequential(torch.nn.AdaptiveMaxPool2d(1), # 100 x 512 x 1 x 1 
                                        torch.nn.Flatten(), # 100 * 512
                                        torch.nn.Dropout(0.2),
                                        torch.nn.Linear(512, number_of_classes) ,
                                        torch.nn.Sigmoid())
  
  def forward(self, xb):
        out = self.conv1(xb)
        out = self.conv2(out)
        out = self.conv3(out)
        out = self.conv4(out)
        out = self.classifier(out)
        return out
    
# Let'sgenerate a fake dataset and pass to it 
input_channels = 3
model = Complex_Model(input_channels = input_channels , number_of_classes = 2 )
# The fake dataset 
import torch 
x = torch.rand(100 , 3  , 32 , 32)
print(model(x).shape)
````

# Training the Model

````python
import torch.nn as nn
dir(nn)
````

````python
# define the loss
import torch
entropy = torch.nn.CrossEntropyLoss()
# define the optimizer
learning_rate = 0.01
optim = torch.optim.Adam(model.parameters() , lr = learning_rate)
````

````python
# The following code collects the loss and accuracy calculated while training the model
def train(num_epochs):
  model.train() # training mode 
  running_loss = 0
  correct_samples = 0
  total = 0

  for inputs , targets in train_loader:
    # pass the inputs to the models
    predictions = model(inputs)
    # calculate the loss
    loss = entropy(predictions , targets)
    # calculate the gradients 
    loss.backward()
    # update the weights 
    optim.step()
    # reset the gradients 
    optim.zero_grad()

    # accumulate the running loss
    running_loss += loss.item()
    _ , scores = predictions.max(dim = 1 )
    correct_samples += (scores ==targets ).sum().item()
    total += targets.size(0)

  # now let's calculate the total running loss and the accuracy
  train_loss  = running_loss / len(train_loader)
  accuracy = (correct_samples / total) * 100
  print("The training loss is  = {}".format(train_loss))
  print("The training accuracy is = {}".format(accuracy))
````

# Evaluate the model using Validation set 

````python
# The following code collects the loss and accuracy calculated while testing the model
def evaluate(num_epochs):
  model.eval()
  running_loss = 0
  correct_samples = 0
  total_size = 0
  with torch.no_grad():
    for inputs , targets in validation_loader:
      predictions = model(inputs)
      loss = entropy(predictions , targets )
      running_loss += loss.item()
      _ , scores = predictions.max(dim = 1)
      correct_samples += (scores == targets).sum().item()
      total_size += targets.size(0)
   # calculate the loss and the accuracy
  validation_loss = running_loss / len(validation_loader)
  validation_accuracy  = (correct_samples / total_size)*100
  print("The validation loss is  = {}".format(validation_loss))
  print("The validation accuracy is ={}".format(validation_accuracy))
````

````python
# Let's call the two functions
epochs = 10
for epoch in range(epochs):
  train(epoch)
  evaluate(epoch)
````

Let's define a helper function predict_image , which it returns the predicteed label for a single image tensor

````python
def predict_image(img , model):
  # convert to batch of 1 , because the image is now 3*32*32 but our model expects N*3*32*32
  xb = img.unsqueeze(0)
  # let's move it to the GPU
  test_img = to_device(data = xb , device = my_device)
  # Get a prediction from the model
  predictions = model(test_img)
  # pick the index with a highest probability
  _ , scores = torch.max(predictions , dim = 1) # but i guess this must be argmax to return the index
  # retrieve the class label
  return test_dataset.classes[scores[0].item()]
````
